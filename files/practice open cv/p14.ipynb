```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
def imshow(title = "Image" , image = None , size = 10):
    w , h = image.shape[0] , image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size , size * aspect_ratio))
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
face = cv2.CascadeClassifier("xmls/haarcascade_frontalface_default.xml")

image = cv2.imread("images/face.jpg")

gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

faces = face.detectMultiScale(gray , 1.3 , 4)

for (x,y,w,h) in faces:
    cv2.circle(image , (x + w//2 , y + h//2) , w//2 , (0,255,0) , 2)
    image = image[y:y+h , x:x+w]

imshow("Face Detection" , image)
```

```python
eye = cv2.CascadeClassifier("xmls/haarcascade_eye.xml")

gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

# faces = eye.detectMultiScale(gray , 1.3 , 4)

eyes = eye.detectMultiScale(gray , 1.3 , 4)
for (x,y,w,h) in eyes:
    cv2.rectangle(image , (x,y) , (x+w , y+h) , (0,255,0) , 2)

imshow("Eye Detection" , image)
```
