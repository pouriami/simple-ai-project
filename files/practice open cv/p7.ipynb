```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def imshow(title = "Image" , image = None , size = 10):
    w , h = image.shape[0] , image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size , size * aspect_ratio))
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

image = cv2.imread("images/mazandaran.jpg")
imshow("Original Image" , image)
```

```python
image
```

```python
M  = np.ones(image.shape , dtype = "uint8") * 100
image.shape
```

```python
added = cv2.add(image , M)
imshow("Added" , added)
added2 = image + M
imshow("Added2" , added2)
```

```python
subtracted = cv2.subtract(image , M)  
imshow("Subtracted" , subtracted)
sub2 = image - M
imshow("Sub2" , sub2)
```

```python
square = np.zeros((300 , 300) , dtype = "uint8")
cv2.rectangle(square , (50 , 50) , (250 , 250) , 255 , -2)
imshow("Square" , square)
elipsis = np.zeros((300 , 300) , dtype = "uint8")
cv2.ellipse(elipsis , (150 , 150) , (150 , 150) , 30 , 0 , 180 , 225 , -1)
imshow("elipsis" , elipsis)
```

```python
And = cv2.bitwise_and(square , elipsis)
imshow("And" , And)
od = cv2.bitwise_or(square , elipsis)
imshow("Or" , od)
Xor = cv2.bitwise_xor(square , elipsis)
imshow("Xor" , Xor)
bitnot = cv2.bitwise_not(square)
imshow("Not" , bitnot)

```
