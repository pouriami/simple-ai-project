draw a shape on open cv

```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def ISF(title='image',image = None , size = 10):
    w , h = image.shape[0] , image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(aspect_ratio*size,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
image = cv2.imread('images/landscape.jpeg')
ISF(image=image)
```

```python
image = np.zeros((512,512,3),np.uint8)
image_gray = np.zeros((512,512),np.uint8)
```

```python
cv2.line(image,(100,100),(100,150),(255,127,0),5)
ISF(image=image)
```

```python
image = np.zeros((512,512,3),np.uint8)
cv2.rectangle(image,(100,100),(200,200),(255,127,0),10)
ISF(image=image)
```

```python
image = np.zeros((512,512,3),np.uint8)
cv2.circle(image,(350,350),100,(255,127,0),-10)
ISF(image=image)
```

```python
image = np.zeros((512,512,3),np.uint8)
pts = np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(image,[pts],isClosed=True,color=(255,127,0),thickness=5)
ISF(image=image)
```

```python
image = np.zeros((2048,2048,3),np.uint8)
outstring = 'hello world'
cv2.putText(image, outstring ,(55,592),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,10,(255,127,0),5)
ISF(image=image)
```
