```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def ISF(title='image',image = None , size = 10):
    w , h = image.shape[0] , image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(aspect_ratio*size,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
image = cv2.imread('images/landscape.jpeg')
ISF(image=image)
```

```python
height , width = image.shape[:2]

quarter_height , quarter_width = height/4 , width/4
```

```python
T = np.float32([[1,0,quarter_width],[0,1,quarter_height]])
```

```python
T
```

```python
img_translation = cv2.warpAffine(image,T,(width,height))
ISF(image=img_translation)
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),90,0.5)
rotaded_image = cv2.warpAffine(image,rotation_matrix,(width,height))
ISF(image=rotaded_image)
```

```python
flipped = cv2.flip(image,1)
ISF(image=flipped)
```
