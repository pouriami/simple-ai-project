```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
def imshow(title = "Image" , image = None , size = 10):
    w , h = image.shape[0] , image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size , size * aspect_ratio))
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

image = cv2.imread("images/mazandaran.jpg" , 0)
imshow("Original Image" , image)
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(image , kernel , iterations = 1)    
imshow("Erosion Image" , erosion)
dilate = cv2.dilate(image , kernel , iterations = 1)
imshow("Dilate Image" , dilate)
opening = cv2.morphologyEx(image , cv2.MORPH_OPEN , kernel)
imshow("Opening Image" , opening)
closing = cv2.morphologyEx(image , cv2.MORPH_CLOSE , kernel)
imshow("Closing Image" , closing)
```

```python
canny = cv2.Canny(image , 50 , 150)
imshow("Canny Image" , canny)
canny = cv2.Canny(image , 10 , 200)
imshow("Canny Image" , canny)
canny = cv2.Canny(image , 200 , 240)
imshow("Canny Image" , canny)
```

```python
def autocanny(image):
    blured_image = cv2.GaussianBlur(image, (5, 5), 0)
    mad_val = np.median(image)
    lower = int(max(0, 0.66 * mad_val))
    upper = int(min(255, 1.33 * mad_val))
    
    edges = cv2.Canny(image, lower, upper)
    return edges

#     
autocanny_image = autocanny(image)
cv2.imshow("Auto Canny Image", autocanny_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
