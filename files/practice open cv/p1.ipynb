```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np
```

```python
cv2.__version__
```

```python
image = cv2.imread('images/flower.jpeg')
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.imshow('image',image_rgb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image_rgb)
plt.show
```

```python
def ISF(title= '',image= None):
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
ISF('hello',image)
```

```python
print(f'height:{image.shape[0]}')
print(f'width:{image.shape[1]}')
print(f'Color components:{image.shape[2]}')
import math
gcd = math.gcd(image.shape[0], image.shape[1])
simplified_height = image.shape[0] // gcd
simplified_width = image.shape[1] // gcd
print(f'dimension: {simplified_height}/{simplified_width}')
```
