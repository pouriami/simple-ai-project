this code about use black and white to image

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

create function for this

```python
def ISF(title='',image = None , size = 10):
    h,w = image.shape[0],image.shape[1]
    aspect_ratio = w/h
    # the bellow code about choose figure size in matplotlib 
    plt.Figure(figsize=(size*aspect_ratio,size))
    # convert BGR to RGB
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('images/landscape.jpeg')
```

```python
ISF("colored image",image=image)
```

to gray scale 

```python
graysc = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
ISF('gray image',image = graysc)
```
