```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
def imshow(title = "Image" , image = None , size = 10):
    w , h = image.shape[0] , image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size , size * aspect_ratio))
    plt.imshow(cv2.cvtColor(image , cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

image = cv2.imread("images/image.png")
imshow("Original Image" , image)
```

```python
ret , tresh1  =  cv2.threshold(image , 127,255,cv2.THRESH_BINARY)
imshow("Binary Threshold" , tresh1)
ret , tresh2  =  cv2.threshold(image , 127,255,cv2.THRESH_BINARY_INV)
imshow("Binary Inverse Threshold" , tresh2)
ret , tresh3  =  cv2.threshold(image , 127,255,cv2.THRESH_TRUNC)    
imshow("Truncated Threshold" , tresh3)
ret , tresh4  =  cv2.threshold(image , 127,255,cv2.THRESH_TOZERO)
imshow("To Zero Threshold" , tresh4)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
thresh = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 5)
imshow("adaptive " , thresh)
```

```python
from skimage.filters import threshold_local

v = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2HSV))[2]
t = threshold_local(v, 29, offset=15, method="gaussian")

thresh = (v > t).astype("uint8") * 255
imshow("Local Threshold" , thresh)
```
